```python
import seaborn as sns
from math import log
```

```python
run_time = [29.77, 24.43, 28.68, 52.64, 75.44, 113.8, 185, 345.6, 569.6, 1173]
run_time = [log(x) for x in run_time]

x_val = [1.25, 2.5, 5, 10, 20, 40, 80, 160, 320, 640]
x_val = list(range(10))

graph = sns.lineplot(x=x_val, y =run_time)
graph.set_xlabel("2$^{scale-factor}$", fontsize = 20)
graph.set_ylabel("time (seconds)", fontsize = 20)
```
